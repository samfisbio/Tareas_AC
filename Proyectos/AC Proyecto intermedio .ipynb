```python
#importamos librerias
import numpy
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
```

```python
def punto():
    coordenadas=[np.random.uniform(-1,1), np.random.uniform(-1,1)]
    return coordenadas[0]**2+coordenadas[1]**2<1, coordenadas
```

```python
def aproxPi(n):
    dentro=0
    Xdentro=[]
    Ydentro=[]
    Xfuera=[]
    Yfuera=[]
    for i in range(n):
        estado, coordenadas = punto()
        if estado:
            dentro+=1
            Xdentro.append(coordenadas[0])
            Ydentro.append(coordenadas[1])
        else:
            Xfuera.append(coordenadas[0])
            Yfuera.append(coordenadas[1])
    return 4*dentro/n, Xdentro, Ydentro, Xfuera, Yfuera
```

```python
n=4444
π=aproxPi(n)

fig=plt.figure(figsize=(5,5))
ax=fig.add_subplot(111)
ax.set_aspect('equal')

circulo = plt.Circle((0, 0), 1, color='0.9', fill=True, zorder=1)
ax.add_artist(circulo)
ax.scatter(π[1],π[2], s=0.2, c='r', marker="o", zorder=2)
ax.scatter(π[3],π[4], s=0.2, c='b', marker="o", zorder=3)

plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.title(f'$n={n:0>9,}$, $\pi \sim {π[0]:0<7.5}$')
plt.show()
```

Con 4444 puntos se estima un vaor de π = .12780

```python
#importamos librerias 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Creamos una función para aproximar el valor de pi a partir del número de puntos

def GraficaPi(num_puntos = 100):
    
    x = np.random.uniform(0, 1, num_puntos)
    y = np.random.uniform(0, 1, num_puntos)
    ConteoPuntos = 0
    
    for i in range(num_puntos):
        
        if x[i]**2 + y[i]**2 <= 1:
            ConteoPuntos += 1
        
    pi_value = 4 * ConteoPuntos / num_puntos
    
    return pi_value
```

```python
GraficaPi(num_puntos = 10000)
```

```python
nr_puntos = []
mc_pi = []

for npuntos in range(1, 1000):
    num_puntos = 100 * npuntos
    nr_puntos.append(num_puntos)
    pi = GraficaPi(num_puntos = num_puntos)
    mc_pi.append(pi)
```

```python
# establecemos los valores y coordenadas
plt.plot(nr_puntos, mc_pi, 'c-', alpha = 0.4)
plt.plot([nr_puntos[0], nr_puntos[-1]], [np.pi, np.pi], 'g--')
xmin, xmax, ymin, ymax = 0, 1.01 * np.amax(nr_puntos), 3.05, 3.25
plt.axis([xmin, xmax, ymin, ymax])
plt.xlabel('Número de puntos')
plt.ylabel('Valor de Pi');
```
